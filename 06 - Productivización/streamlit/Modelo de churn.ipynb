Implementa un modelo de churn utilizando las siguientes variables de entrada:  
* `gender`  
* `PaymentMethod`  
* `MonthlyCharges`  
* `tenure`

```python
df_churn = df_churn[['gender','PaymentMethod','MonthlyCharges','tenure','Churn']].copy()
df_churn
```

```python
df = df_churn.copy()
df.fillna(0,inplace=True)
```

```python
# Codificar las variables
encode = ['gender','PaymentMethod']

for col in encode:
    dummy = pd.get_dummies(df[col],prefix=col)
    df = pd.concat([df,dummy],axis=1)
    del df[col]
```

```python
df
```

```python
import numpy as np
df['Churn'] = np.where(df['Churn']=='Yes',1,0)
```

```python
df
```

```python
X = df.drop('Churn',axis=1)
y = df['Churn']
```

```python
# DivisiÃ³n train-test
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.2)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier()
clf.fit(X_train,y_train)
```

```python
y_predict = clf.predict(X_test)
```

```python
y_predict
```

```python
from sklearn.metrics import classification_report, confusion_matrix
print(classification_report(y_test,y_predict))
print(confusion_matrix(y_test,y_predict))
```

```python
import pickle
pickle.dump(clf, open('churn_clf.pkl', 'wb'))
```
